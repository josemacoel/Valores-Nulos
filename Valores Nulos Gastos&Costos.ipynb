# 2020

```python
df2020= pd.read_excel('GastosCostos2023_adecuado.xlsx', sheet_name="2020")
df2020
```

```python
valores_nulos2020=df2020.isnull().sum()
valores_nulos2020
```

```python
df2020.info()
```

```python
#reemplazando los valores nulos de la columna FOLIO por valores en 0
df2020["FOLIO"].fillna("", inplace=True)
#Sustituyendo los conceptos de pagos que no aparecen por Gasto Omitido
df2020['GASTO'].fillna('OMITIDO', inplace=True)
#Sustituyendo los conceptos de TC que no aparecen por 0
df2020['TC'].fillna(0, inplace=True)
#Sustituyendo los conceptos de IMPORTE que no aparecen por 0
df2020['IMPORTE'].fillna(0, inplace=True)
#Sustituyendo los conceptos de IVA que no aparecen por 0
df2020['IVA'].fillna(0, inplace=True)
#Sustituyendo los conceptos de POLIZA que no aparecen por OMITIDA
df2020['POLIZA'].fillna("OMITIDA", inplace=True)
#Sustituyendo los conceptos de TIPO que no aparecen por OMITIDA
df2020['TIPO'].fillna("OMITIDA", inplace=True)
```

```python
valores_nulos2020=df2020.isnull().sum()
valores_nulos2020
```
