# **Detalle precios y productos fabricados**

```python
#Carga desde un archivo .xlsx sin indice
precios = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
precios.head()
```

```python
#Checar si hay valores nulos por columna
valores_nulos=precios.isnull().sum()
valores_nulos
```

```python
#como se puede observar existen dos valores nulos porque no tiene el nombre de el vendedor,
#tiene un valor en blanco y no es posible saber el nombre. Por lo que la mejor opción es remplazar
#esos datos por una categoria "vendedor anonimo"
```

```python
precios['NOMBRE_VENDEDOR'].fillna(value='VENDEDOR ANONIMO', inplace=True)
```

```python
#Checar si hay valores nulos por columna
valores_nulos=precios.isnull().sum()
valores_nulos
```
